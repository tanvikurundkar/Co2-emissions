# Cleaning the data:

```python
df_1=df.loc[df['year']>1990] 
df_new1=df_1.fillna(df['co2_per_gdp'].mean()) 
df_new2=df_new1.fillna(df['co2'].mean()) 
df_new3=df_new2.fillna(df['co2_per_capita'].mean()) 
df_new4=df_new3.fillna(df['cumulative_co2'].mean()) 
df_new5=df_new4.fillna(df['coal_co2'].mean()) 
df_new6=df_new5.fillna(df['cement_co2'].mean()) 
df_new=df_new6.fillna(df['oil_co2'].mean()) 
#df_new=df_new7.fillna(df['gdp'].median()) 
df_new.head()
```

We have replaced the Null values with the average value of each column using the ".mean" function so that we do not receive inaccurate results. Moreover, we have considered years only after 1990 since the years before contain more null values which would adversely affect our results.

# The trend of the world's co2 emissions from 1990 to 2019 per decade

```python
i= df_new['year'].min()
xlabel_list=[]
decade_co2_lst=[]
while i<(df_new['year'].max())+1:
  decade_co2 = 0
  for j in range(10):
    year_x=i+j
    decade_co2 = decade_co2 + df_new.loc[df_new['year']==year_x,'co2'].sum()
  decade_co2_lst.append(np.round(decade_co2.mean(),decimals=1))
  xlabel_list.append(str(i)+'-'+str(i+10))
  i+=10
fig = plt.figure(figsize=(8,6))
plt.xlabel('Decade')
plt.ylabel('Avg CO2 Emission')
plt.title('The trend of the world\'s co2 emissions from 1990 to 2019 per decade')
plt.bar(x=xlabel_list,height=decade_co2_lst)
plt.show()

    
```

**Therefore, the average CO2 emissions have been increasing gradually until they reached 350,000 million tonnes in the latest decade.**

# Which country had the highest co2 emissions and in which year ??

```python
co2_max=df_new['co2'].max()
co2_max_mask=df_new[df_new['co2'] == co2_max].index[0]
max_co2_country=df_new.loc[co2_max_mask,'country']
max_co2_year=df_new.loc[co2_max_mask,'year']
print('*****************************************************************************\n')
print(f'{max_co2_country} emitted the most co2 i.e. {co2_max} million tonnes in the year {max_co2_year} \n')
print('*****************************************************************************')
```

# How does population of a country affect its CO2 emissions ??

```python
country_list =input('Enter countries to compare population vs co2(split by a comma): ').split(',')
# Subplots
fig,ax = plt.subplots(len(country_list),figsize=(15,20))
axis_number=0
for country in country_list:
  country_population=df_new.loc[df_new['country']==country,'population'] 
  country_co2=df_new.loc[df_new['country']==country,'co2']
  ax[axis_number].plot(country_population.values,country_co2.values)
  ax[axis_number].title.set_text(country) 
  ax[axis_number].set_xlabel('Population in 100 millions')
  ax[axis_number].set_ylabel('Co2 Emissions in million tonnes')
  axis_number+=1 
plt.show() 
# Singular Graph
figure = plt.figure(figsize=(20,15)) 
for country in country_list:
  country_population=df_new.loc[df_new['country']==country,'population'] 
  country_co2=df_new.loc[df_new['country']==country,'co2']
  plt.plot(country_population.values,country_co2.values,label=country)
plt.legend() 
plt.xlabel('Population')
plt.ylabel('Co2 Emissions') 
plt.show()
```

**For majority of the countries,there is a positive correlation between population and CO2 emissions; as the population rose, the CO2 emissions increased as well. (directly proportional)**

**For comparing the effect on population between multiple countries, the user must enter countries with similar population, in order to get a better idea of how it is working.**

# The trend in CO2 emissions in  a given country since 1990

```python
ctry=str(input('Enter the country: '))
ctry_years=df_new.loc[df_new['country']==ctry,'year'] 
ctry_co2_trend=df_new.loc[df_new['country']==ctry,'cumulative_co2'] 
fig = plt.figure(figsize=(8,6)) 
plt.xlabel('Year')
plt.ylabel('Co2 Emissions') 
plt.title('The trend in CO2 emissions in {} since 1990'.format(ctry))
plt.plot(ctry_years.values,ctry_co2_trend.values)
plt.show()
```

**As we used the cumulative frequency for this graph, there will always be a positive trend in the curve, however, the graph will be different for every country as the rate at which the co2 changes is different for each of them.**

# Drivers of CO2 emissions in a country

```python
country_input=str(input('Enter the name of the country: '))
country_data=df_new.loc[df_new['country']==country_input,['year',"coal_co2","cement_co2", "oil_co2"]] 
coal_average = round(country_data["coal_co2"].mean(),2)
cement_average = round(country_data["cement_co2"].mean(),2)
oil_average = round(country_data["oil_co2"].mean(),2)
print('\n*****************************************************************************')
if coal_average>cement_average and coal_average>oil_average:
    print("COAL is the biggest cause of CO2 emissions in",country_input)
elif cement_average>coal_average and cement_average>oil_average:
    print("CEMENT is the biggest cause of CO2 emissions in",country_input)
elif oil_average>coal_average and oil_average>cement_average:
    print("OIL is the biggest cause of CO2 emissions in",country_input)
else:
    print("The major cause of CO2 emissions in ", country_input, " is unidentifiable")
print('*****************************************************************************') 
```

We have used "if" statements to compare the 3 main causes of CO2, coal, oil and cement. The else statement is used to prevent the user from entering the wrong input.

```python
country_input=str(input('Enter the name of the country: '))
sizes_lst=[]
country_data=df_new.loc[df_new['country']==country_input,['co2','total_ghg','methane','nitrous_oxide']] 
co2_average = round(country_data["co2"].mean(),2)
methane_average = round(country_data["methane"].mean(),2)
nitrous_oxide_average = round(country_data["nitrous_oxide"].mean(),2)
sizes_lst.append(co2_average)
sizes_lst.append(methane_average)
sizes_lst.append(nitrous_oxide_average)
fig = plt.figure(figsize=(8,6))
plt.title('Gases emitted as a part of total greenhouse gases')
plt.pie(sizes_lst,labels=['co2','methane','nitrous oxide'],autopct='%1.0f%%' ) 
plt.show()
```

These programs will help the user understand the primary cause of the CO2 emissions in the given country. Therefore, the country will be aware of the driver of their co2 emissions and can begin using that resource more judiciously.
